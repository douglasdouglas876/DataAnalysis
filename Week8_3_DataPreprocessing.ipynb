### creating dummy variables

```python
cols = ['Embarked']
pd.get_dummies(df[cols])
```

```python
dummies = []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
    dummies.append(pd.get_dummies(df[col]))
dummies
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
```

```python
df = pd.concat((df,titanic_dummies), axis=1)
```

```python
df = df.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
```

### more on interpolate

```python
import pandas as pd
df = pd.DataFrame({"A":[12, 4, 5, None, 1],
                   "B":[None, 2, 54, 3, None],
                   "C":[20, 16, None, 3, 8],
                   "D":[14, 3, None, None, 6]})
```

```python
df
```

```python
df.interpolate(method ='linear', limit_direction ='forward')
```

```python
df.interpolate(method ='linear', limit_direction ='backward', limit = 1) 
```

interpolate the missing values in the backward direction using linear method and putting a limit on maximum number of consecutive Na values that could be filled.
